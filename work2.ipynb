```python
import sympy as sym
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid()


def derivative(func, a, h):
    return (func(a + h) - func(a - h)) / (2 * h)
h = 1e-5
x= np.linspace(-10, 10, 100)
func=x**3+x**2+5*x+4
dfunc= derivative(lambda x :x**3+x**2+5*x+4 , x, h)
ddfunc= derivative(lambda x :3*x**2+2*x+5 , x, h)

ax.plot(x, func, color="blue")
ax.plot(x, dfunc, color="red")
ax.plot(x, ddfunc, color="green")
```
