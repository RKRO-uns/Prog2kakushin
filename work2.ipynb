```python
import numpy as np
import matplotlib.pyplot as plt

posi= np.array([[-10,-20,-25,-16,-10,0,-8,30,16,23,11,-10,11,23,50,58,59,57.5,55],[150,130,128,113,108,75,45,48,80,103,118,108,118,103,130,114,105,100,93]])
posi1=np.array([[8],[77.5]])

posi2=np.array([[0,-100,-85,-55,-63,-42,-39,-23,0],[0,25,75,125,80,48,25,13,0]])
posi3=np.array([[-55,-65,-55,-56,-55,-45,-26,26,70],[125,118,125,132,125,121.5,135,134,140]])
posi4=np.array([[-26,-20,5],[135,100,81]])

posi5=np.array([[0,40,70,95,84,95,90],[0,50,70,81,95,81,14]])
posi6=np.array([[95,105],[81,18]])

print("星座の名前を当てろ！！")
print("全部で３問あるよ")
print("３回答えるチャンスがあり、１回ミスするごとにヒントがあるよ。(回答例　いて座)")
print("---------------------------------------------------------------------------------")
print("第１問")
n=0
while True:
  if n<2:
    if n==1:
      plt.plot(posi[0, :], posi[1, :],color="blue")
      plt.plot(posi1[0, :], posi1[1, :],color="blue")
    plt.plot(posi[0,:], posi[1,:], '.', color='red')
    plt.plot(posi1[0,:], posi1[1,:], '.', color='red')
    plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False, bottom=False, left=False, right=False, top=False)
    plt.axis('equal')
    plt.show()
  a=input("何座？：")
  if a=="オリオン座" or a=="Orion" or a=="おりおん座":
    print("正解！")
    break
  if n>=2:
    print("答えは「オリオン座」でした。")
    break
  if n==0:
    print("ヒント：線でつないでみました。")
  elif n==1:
    print("ヒント：オ○○〇座")
  n+=1
plt.cla()

print("---------------------------------------------------------------------------------")
print("第２問")
t=0
while True:
  if t<2:
    if t==1:
      plt.plot(posi2[0, :], posi2[1, :],color="blue")
      plt.plot(posi3[0, :], posi3[1, :],color="blue")
      plt.plot(posi4[0, :], posi4[1, :],color="blue")
    plt.plot(posi2[0,:], posi2[1,:], '.', color='red')
    plt.plot(posi3[0,:], posi3[1,:], '.', color='red')
    plt.plot(posi4[0,:], posi4[1,:], '.', color='red')
    plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False, bottom=False, left=False, right=False, top=False)
    plt.axis('equal')
    plt.show()
  b=input("何座？：")
  if b=="、みずがめ座"  or a=="水瓶座":
    print("正解！")
    break
  if t>=2:
    print("答えは「みずがめ座」でした。")
    break
  if t==0:
    print("ヒント：線でつないでみました")
  elif t==1:
    print("ヒント：誕生日が１月２０日から２月１８日生まれの誕生星座です。")
  t+=1
plt.cla()

print("---------------------------------------------------------------------------------")
print("第３問")
k=0
while True:
  if k<2:
    if k==1:
      plt.plot(posi5[0, :], posi5[1, :],color="blue")
      plt.plot(posi6[0, :], posi6[1, :],color="blue")
    plt.plot(posi5[0,:], posi5[1,:], '.', color='red')
    plt.plot(posi6[0,:], posi6[1,:], '.', color='red')
    plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False, bottom=False, left=False, right=False, top=False)
    plt.axis('equal')
    plt.show()
  c=input("何座？：")
  if c=="いっかくじゅう座" or c=="一角獣座":
    print("正解！")
    break
  if k>=2:
    print("答えは「いっかくじゅう座」でした。")
    break
  if k==0:
    print("ヒント：線でつないでみました。")
  elif k==1:
    print("ヒント：ユニコーンをモチーフとした星座です。")
  k+=1
```
